## 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
# import cleaned products csv
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_wrangled.csv'), index_col = False)
```

```python
df_prods.head()
```

```python
# clean up dataframes to get rid of blank columns: find column names to delete
df_ords_prods.columns
```

```python
# clean up dataframes to get rid of blank columns
df_ords_prods2 = df_ords_prods.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', '_merge'])
```

```python
df_ords_prods2.head()
```

```python
# clean up df_prods
df_prods2 = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods2.head()
```

```python
# merging dataframes on product_id key column as it is the only column that is in both dataframes
df_orders_products = df_ords_prods2.merge(df_prods2, on = 'product_id', indicator = True)
```

```python
df_orders_products
```

## 6. Confirm the results of the merge using the merge flag.

```python
df_orders_products['_merge'].value_counts()
```

## 7. Export the newly created dataframe as orders_products_merged in a suitable format (taking into consideration the size).

```python
# using pkl because of the size of the new dataframe
df_orders_products.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```
